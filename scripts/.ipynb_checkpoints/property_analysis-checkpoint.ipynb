```python
def get_concept_label_dict(prop_dict):
    
    concept_label_dict = dict()
    labels_pos =  ['all', 'all-some', 'some', 'few-some']
    labels_neg = = ['few']
    for c, d in prop_dict.items():
        ml_label = d['ml_label']
        if ml_label in labels_pos:
            concept_label_dict[c] = 'pos'
        elif ml_label in labels_neg:
            concept_label_dict[c] = 'neg'
    return concept_label_dict



    
def get_closest_furthest(distance_concept_list,concept_pair_dict, mode='first_negative'):
    
    target = None
    if mode == 'first_negative':
        rank = 1
        target_labels = ['few']
    elif mode == 'last_positive':
        rank = len(distance_concept_list)
        target_labels = ['all', 'all-some', 'some', 'few-some']
        distance_concept_list = distance_concept_list[::-1]
    
    for dist, w_form, w in distance_concept_list:
        pair = concept_pair_dict[w]
        label = pair.ml_label
        if label in target_labels and target is None:
            target = dict()
            target['word'] = w
            target['label'] = label
            target['cosine'] = dist
            target['rank'] = rank
        if mode == 'first_negative':
            rank +=1
        elif mode == 'last_positive':
            rank -= 1
    return target



def get_mapped_label_df(distance_concept_list, concept_pair_dict):
    mapping = dict()

    mapping = dict()
    mapping['pos'] = 1
    mapping['neg'] = 0
    
    pos_labels = ['all', 'all-some', 'some', 'few-some']
    neg_labels = ['few']
    

    dist_label_dicts = []
    for cos, w_form, concept in distance_concept_list:
        label_ml = concept_pair_dict[concept].ml_label
        if label_ml in pos_labels:
            label = 'pos'
        elif label_ml in neg_labels:
            label = 'neg'
        else:
            label = 'no-label'
        d = dict()
        if label in mapping:
            d['cos'] = cos
            d['label'] = label
            d['num_label'] = mapping[label]
            d['concept'] =concept
            dist_label_dicts.append(d)
    return pd.DataFrame(dist_label_dicts)


def get_cosine_overview(prop_dict, model_name, model_dict):
    model = model_dict['model']
    concept_pair_dict = dict([(p[1], pair) for p, pair in prop_set.pairs.items()])
    concepts = list(concept_pair_dict.keys())
    concepts_clean = []
    positive_examples = get_positive_examples(prop_set)
    centroid, oov = utils.get_centroid(positive_examples, model)
    distance_concept_list, oov = utils.get_distances_to_centroid(centroid, concepts, model)
    distance_concept_list= sorted(distance_concept_list, reverse = True)
    word_closest = distance_concept_list[0][1]
    word_furthest = distance_concept_list[-1][1]
    d_overview= dict()
    #d_overview['property'] = prop
    #d_overview['# words']=  len(distance_concept_list)
    d_overview[(model_name, 'closest neg')] = get_closest_furthest(distance_concept_list,
                                                     concept_pair_dict, 
                                                     mode='first_negative')['rank']
    d_overview[(model_name, 'furthest pos')] = get_closest_furthest(distance_concept_list,
                                                      concept_pair_dict, 
                                                      mode='last_positive')['rank']
    #d_overview['oov'] = oov
    df = get_mapped_label_df(distance_concept_list, concept_pair_dict)
    #prop_distributions[prop] = df
    spearm_corr = stats.spearmanr(df['num_label'], df['cos'])
    rho, p = spearm_corr

    d_overview[(model_name, 'spearmanrho')] = rho
    return d_overview, df


def get_distribution(data_set, model_names):
    
    overviews = defaultdict(dict)
    prop_distributions = []
    for model_name in model_names:
        model_dict = load_model(model_name)
        for prop, prop_set in data_set.prop_sets.items():
            d_overview, df_prop = get_cosine_overview(prop_set, model_name, model_dict)
            overviews[prop].update(d_overview)
    return overviews, prop_distributions
```

```python
def get_lexical_df(prop_set, vocab_by_words, target_label):

    set_data = []
    pos_labels = ['all', 'all-some', 'some', 'few-some']
    neg_labels = ['few']
    if target_label == 'pos':
        target_labels = pos_labels
    elif target_label == 'neg':
        target_labels = neg_labels
    else:
        print('target label not valid')
    for p, pair in prop_set.pairs.items():
        p, c = p
        ml_label = pair.ml_label
        if ml_label in target_labels:
            lex_data = vocab_by_words[c]
            d = dict()
            if len(lex_data) == 1:
                lex_data = lex_data[0]
                d['concept'] = c
                d['freq'] = lex_data['wiki_frequency']
                d[('wiki_corpus', 'freq')] = lex_data[('wiki_corpus', 'freq')]
                d[('giga_corpus', 'freq')] = lex_data[('giga_corpus', 'freq')]
                d['wn_synsets'] = lex_data['wn_synsets']
                d['ambiguity'] = lex_data['polysemy_type']
                d['conc'] = lex_data['conc']
                d['fam'] = lex_data['fam']
                d['imag'] = lex_data['imag']
                set_data.append(d)
    return set_data

        
```

```python

metrics = [('wiki_corpus', 'freq'), ('giga_corpus', 'freq'),'conc', 'fam', 'imag', 'wn_synsets']
labels = ['pos', 'neg']
lexical_data_dicts = []
prop_data_dicts = defaultdict(dict)
for prop, prop_set in data_set.prop_sets.items():
    for label in labels:
        set_data =  get_lexical_df(prop_set, vocab_by_words, label)
        df = pd.DataFrame(set_data)
        df_dummies = df
        d = dict()
        d['property'] = (prop, label)
        for m in metrics:
            # convert to int or float, empty strings ==> nan ('coerce')
            df_dummies[m] = pd.to_numeric(df_dummies[m], errors='coerce')
            d[(m, 'mean')] = df_dummies[m].mean(axis = 0)
            d[(m, 'median')] = df_dummies[m].median(axis = 0)
            d['label'] = label
            #d[(m, 'std')] = df_dummies[m].std(axis = 0) 
            prop_data_dicts[m][(prop, label)] = df_dummies[m]
        #prop_freq[prop] = freq
        lexical_data_dicts.append(d)
```

```python
def get_properties():
    properties = []
    for path in os.listdir('../data/aggregated/'):
        prop = path.split('.')[0]
        if 'female-' not in prop and prop != '':
            properties.append(prop)
    return properties

def load_prop_set(prop):
    
    path = f'../data/aggregated/{prop}.json'
    with open(path) as infile:
        prop_dict = json.load(infile)
    return prop_dict

def get_relation_distribution(properties):
    rel_dist_dict = defaultdict(dict)
    pos_labels = ['all', 'all-some', 'some', 'few-some']
    for prop in properties:
        prop_dict = load_prop_set(prop)
        rel_prop_pos = defaultdict(list)
        #rel_cnt = Counter()
        for c, d in prop_dict.items():
            prop_type = d['property_type']
            ml_label = d['ml_label']
            if ml_label in pos_labels:
                relations = d['relations']
                for rel, prop_pos in relations.items():
                    #rel_prop_pos[rel].append(prop_pos)
                    if prop_pos > 0.5:
                        rel_prop_pos[rel].append(1)
                    else:
                         rel_prop_pos[rel].append(0)
#                     #rel_prop_pos[rel].append(prop_pos)
#                         rel_prop_pos[rel].append(1)
#                     else:
#                         rel_prop_pos[rel].append(0)
                    #rel_cnt[rel] += 1
        for rel, prop_pos in rel_prop_pos.items():
            d = dict()
            #n = rel_cnt[rel]
            d[(prop_type, prop)] = sum(prop_pos)/len(prop_pos)
            #d[(prop, 'median')] = median(prop_pos)
            #d[(prop, 'n')] = len(prop_pos)
            rel_dist_dict[rel].update(d)
    return rel_dist_dict




properties =  get_properties()

relations = [
    'implied_category',     
    'typical_of_concept', 
    'typical_of_property', 
    'affording_activity', 
    'afforded_usual', 
    'afforded_unusual',
    'variability_limited', 
    'variability_open', 
    'rare',
    'unusual', 'impossible', 'creative'
]
rel_dist_dict = get_relation_distribution(properties)

df = pd.DataFrame(rel_dist_dict).T.round(2)
columns = df.columns
columns_sorted = sorted(columns)
df = df.reindex(relations)[columns_sorted].T
df
#print(df.fillna('-').to_latex())
```

# Get relation configurations


```python
import os
from collections import Counter
import numpy as np
```

```python

### old
import os
from collections import Counter, defaultdict
import csv
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)


def get_properties():
    properties = []
    for path in os.listdir('../data/aggregated/'):
        prop = path.split('.')[0]
        if 'female-' not in prop and prop != '':
            properties.append(prop)
    return properties


def relations_to_file(properties):
    path_dir = '../data/relations'
    os.makedirs(path_dir, exist_ok=True)
    relation_pair_dict = defaultdict(set)
    for prop in properties:
        prop_dict = analyze_evidence.load_prop_data(prop)
        rels_evidence = {'typical_of_property', 'variability_limited',  'afforded_usual', 'affording_activity'}
        rels_no_evidence = {'typical_of_concept', 'afforded_unusual', 'implied_category', 'variability_open'}

        for c, d in prop_dict.items():
            # sort by lables
            ml_label = d['ml_label']
            if ml_label in {'all', 'some', 'all-some', 'few-some'}:
                l = 'pos'
            elif ml_label in {'few'}:
                l = 'neg' 
            relation_pair_dict[(l,)].add((prop, c))

            # sort by subset of instances
            if ml_label in {'all', 'all-some'}:
                l_qu = 'all'
            elif ml_label in {'some', 'few-some'}:
                l_qu = 'some'
            elif ml_label in {'few'}:
                l_qu = 'few'
            relation_pair_dict[(l_qu,)].add((prop, c))

            # sort relations:
            rel_dict = d['relations']
            prop_rel_dict = defaultdict(list)
            for rel, p in rel_dict.items():
                prop_rel_dict[p].append(rel)
            pos_props = sorted(list(prop_rel_dict.keys()), reverse=True)
            relations = []
            for p in pos_props:
                if p > 0.5:
                    relations.extend(prop_rel_dict[p])
            relations = tuple(relations)
            relation_pair_dict[relations].add((prop, c))

            # sort relations according to hypotheses
            relations = [rel for rel, p in rel_dict.items() if p > 0.5]
            rel_ev = [rel for rel in relations if rel in rels_evidence]
            rel_no_ev  = [rel for rel in relations if rel in rels_no_evidence]

            if len(rel_ev) > 0:
                hyp = 'evidence'
            elif len(rel_no_ev) > 0:
                hyp = 'no_evidence_pos'
            else:
                hyp = 'no_evidence_neg'
            relation_pair_dict[(hyp, )].add((prop, c))

        # to file
        for rel, pairs in relation_pair_dict.items():
            name = '-'.join(rel)
            path = f'{path_dir}/{name}.txt'
            with open(path,  'w') as outfile:
                for prop, c in pairs:
                    outfile.write(f'{prop},{c}\n')
                    
# properties = get_properties()                   
# relations_to_file(properties)
```
